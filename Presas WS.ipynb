```python
nombre = []
fase = []
coordenadas = []
for i in range(1227):
    navegador.get("http://www.seprem.es/ficha.php?idpresa="+str(i)) 
    source = navegador.page_source
    soup = BeautifulSoup(source)
    nombre.append(soup.find_all("td", style="font-weight:bold;")[2].get_text())
    fase.append(soup.find_all("td", style="font-weight:bold;")[6].get_text())
    coordenadas.append(soup.find_all("td", style="font-weight:bold;")[18].get_text())
```

```python
presas=pd.DataFrame({"Nombre": nombre,"Fase":fase,"Coordenadas":coordenadas})
presas.head()
```
